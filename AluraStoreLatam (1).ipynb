### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

# 2. Ventas por categoría

```python
print('Tienda 1')
print(tienda["Categoría del Producto"].value_counts().head(3))
print('Tienda 2')
print(tienda2["Categoría del Producto"].value_counts().head(3))
print('Tienda 3')
print(tienda3["Categoría del Producto"].value_counts().head(3))
print('Tienda 4')
print(tienda4["Categoría del Producto"].value_counts().head(3))

```

```python
cat1 = tienda["Categoría del Producto"].value_counts().head(5)
cat2 = tienda2["Categoría del Producto"].value_counts().head(5)
cat3 = tienda3["Categoría del Producto"].value_counts().head(5)
cat4 = tienda4["Categoría del Producto"].value_counts().head(5)

df_categorias = pd.DataFrame({
    'Tienda 1': cat1,
    'Tienda 2': cat2,
    'Tienda 3': cat3,
    'Tienda 4': cat4
})

df_categorias.fillna(0, inplace=True)  # Rellena con 0 donde no hay datos
df_categorias = df_categorias.astype(int)

df_categorias.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Top 3 categorías más vendidas por tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.legend(title='Tiendas')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
print('Tienda 1 - Promedio de calificación:')
print(tienda["Calificación"].mean().round(2))

print('Tienda 2 - Promedio de calificación:')
print(tienda2["Calificación"].mean().round(2))

print('Tienda 3 - Promedio de calificación:')
print(tienda3["Calificación"].mean().round(2))

print('Tienda 4 - Promedio de calificación:')
print(tienda4["Calificación"].mean().round(2))
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
print('Tienda 1')
conteo1 = tienda["Producto"].value_counts()
print("Producto más vendido:", conteo1.idxmax(), "-", conteo1.max(), "ventas")
print("Producto menos vendido:", conteo1.idxmin(), "-", conteo1.min(), "ventas\n")

# Tienda 2
print('Tienda 2')
conteo2 = tienda2["Producto"].value_counts()
print("Producto más vendido:", conteo2.idxmax(), "-", conteo2.max(), "ventas")
print("Producto menos vendido:", conteo2.idxmin(), "-", conteo2.min(), "ventas\n")

# Tienda 3
print('Tienda 3')
conteo3 = tienda3["Producto"].value_counts()
print("Producto más vendido:", conteo3.idxmax(), "-", conteo3.max(), "ventas")
print("Producto menos vendido:", conteo3.idxmin(), "-", conteo3.min(), "ventas\n")

# Tienda 4
print('Tienda 4')
conteo4 = tienda4["Producto"].value_counts()
print("Producto más vendido:", conteo4.idxmax(), "-", conteo4.max(), "ventas")
print("Producto menos vendido:", conteo4.idxmin(), "-", conteo4.min(), "ventas\n")
```

# 5. Envío promedio por tienda

```python
print('Tienda 1 - Frecuencia de costos de envío (mayores a 0):')
envios_filtrados = tienda[tienda["Costo de envío"] > 0]
print(envios_filtrados["Costo de envío"].value_counts().mean().round(2))


print('Tienda 2 - Frecuencia de costos de envío (mayores a 0):')
envios_filtrados = tienda2[tienda2["Costo de envío"] > 0]
print(envios_filtrados["Costo de envío"].value_counts().mean().round(2))



print('Tienda 3 - Frecuencia de costos de envío (mayores a 0):')
envios_filtrados = tienda3[tienda3["Costo de envío"] > 0]
print(envios_filtrados["Costo de envío"].value_counts().mean().round(2))


print('Tienda 4 - Frecuencia de costos de envío (mayores a 0):')
envios_filtrados = tienda4[tienda4["Costo de envío"] > 0]
print(envios_filtrados["Costo de envío"].value_counts().mean().round(2))
```

```python
Ventas_totales_tienda_1 = tienda.shape[0]
print(Ventas_totales_tienda_1)
Ventas_totales_tienda_2 = tienda2.shape[0]
print(Ventas_totales_tienda_2)
Ventas_totales_tienda_3 = tienda3.shape[0]
print(Ventas_totales_tienda_3)
Ventas_totales_tienda_4 = tienda4.shape[0]
print(Ventas_totales_tienda_4)
```

```python

categorias = ['Ventas totales tienda 1', 'Ventas totales tienda 2', 'Ventas totales tienda 3', 'Ventas totales tienda 4']
ventas = [Ventas_totales_tienda_1, Ventas_totales_tienda_2, Ventas_totales_tienda_3, Ventas_totales_tienda_4]

# Crear gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(ventas, labels=categorias, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Ventas totales por tienda')
plt.axis('equal')
plt.show()
```

```python
Ventas_totales_tienda_1_lugar = tienda["Lugar de Compra"].value_counts()
Ventas_totales_tienda_2_lugar = tienda2["Lugar de Compra"].value_counts()
Ventas_totales_tienda_3_lugar = tienda3["Lugar de Compra"].value_counts()
Ventas_totales_tienda_4_lugar = tienda4["Lugar de Compra"].value_counts()
```

```python
ventas_por_lugar = (
    Ventas_totales_tienda_1_lugar
    + Ventas_totales_tienda_2_lugar
    + Ventas_totales_tienda_3_lugar
    + Ventas_totales_tienda_4_lugar
).fillna(0).astype(int)
```

```python
print("Ventas totales por lugar (todas las tiendas combinadas):")
print(ventas_por_lugar.sort_values(ascending=False))
```

```python
plt.figure(figsize=(10, 6))
ventas_por_lugar.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='black')

# Personalización
plt.title("Ventas totales por ciudad (todas las tiendas)", fontsize=14)
plt.xlabel("Ciudad", fontsize=12)
plt.ylabel("Cantidad de ventas", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

```python
dinero_generado_tienda_1 = tienda["Precio"].sum()
dinero_generado_tienda_2 = tienda2["Precio"].sum()
dinero_generado_tienda_3 = tienda3["Precio"].sum()
dinero_generado_tienda_4 = tienda4["Precio"].sum()
```

```python
print(f'Dinero generado tienda 1: ${dinero_generado_tienda_1:,.2f}')
print(f'Dinero generado tienda 2: ${dinero_generado_tienda_2:,.2f}')
print(f'Dinero generado tienda 3: ${dinero_generado_tienda_3:,.2f}')
print(f'Dinero generado tienda 4: ${dinero_generado_tienda_4:,.2f}')



```

```python

# Datos de ejemplo
categorias = ['Dinero total tienda 1', 'Dinero total tienda 2', 'Dinero total tienda 3', 'Dinero total tienda 4']
ventas = [dinero_generado_tienda_1, dinero_generado_tienda_2, dinero_generado_tienda_3, dinero_generado_tienda_4]


# Gráfico de líneas
plt.figure(figsize=(8, 5))
plt.plot(categorias, ventas, marker='o', linestyle='-', color='green')

# Personalización
plt.title('Dinero total generado por cada tienda', fontsize=14)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Dinero generado ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
